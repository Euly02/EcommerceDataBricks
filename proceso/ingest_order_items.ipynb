#Imports

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

#Widgets

```python
dbutils.widgets.removeAll()
dbutils.widgets.text("storageName", "adlsecommerceproject")
dbutils.widgets.text("container","raw")
dbutils.widgets.text("catalog","catalog_ecommerce")
dbutils.widgets.text("schema","bronze")
```

#Constantes

```python
storageName = dbutils.widgets.get("storageName")
container = dbutils.widgets.get("container")
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
```

#Estructura

```python
items_schema = StructType(fields=[
    StructField("order_id", StringType(), False),
    StructField("order_item_id", StringType(), False),
    StructField("product_id", StringType(), False),
    StructField("seller_id", StringType(), False),
    StructField("shipping_limit_date", StringType(), False),
    StructField("price", DecimalType(10,2), False),
    StructField("freight_value", DecimalType(10.2), False),
])
```

#Leer CSV

```python
df_items = spark.read.format("csv").option('header', True).schema(items_schema).load(f"abfss://{container}@{storageName}.dfs.core.windows.net/orders_items")
```

#Guardar en tabla

```python
df_items.write.mode("overwrite").saveAsTable(f"{catalog}.{schema}.orders_items")
```
