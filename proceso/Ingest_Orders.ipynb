#Imports

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

#Widgets

```python
dbutils.widgets.removeAll()
dbutils.widgets.text("storageName", "adlsecommerceproject")
dbutils.widgets.text("container","raw")
dbutils.widgets.text("catalog","catalog_ecommerce")
dbutils.widgets.text("schema","bronze")
```

#Constantes

```python
storageName = dbutils.widgets.get("storageName")
container = dbutils.widgets.get("container")
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
```

#Estructura

```python
orders_schema = StructType(fields=[
    StructField("order_id", StringType(), False),
    StructField("customer_id", StringType(), False),
    StructField("order_status", StringType(), False),
    StructField("order_purchase_timestamp", TimestampType(), False),
    StructField("order_approved_at", TimestampType(), False),
    StructField("order_delivered_carrier_date", TimestampType(), False),
    StructField("order_delivered_customer_date", TimestampType(), False),
    StructField("order_estimated_delivery_date", TimestampType(), False),
])
```

#Leer el archivo

```python
df_orders = spark.read.option('header', True).schema(orders_schema).csv(f"abfss://{container}@{storageName}.dfs.core.windows.net/orders")
```

#Guardar la tabla

```python
df_orders.write.mode("overwrite").saveAsTable(f"{catalog}.{schema}.orders")
```
