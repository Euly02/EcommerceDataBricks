#Import

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

#Widgets

```python
dbutils.widgets.removeAll()
dbutils.widgets.text("storageName", "adlsecommerceproject")
dbutils.widgets.text("container","raw")
dbutils.widgets.text("catalog","catalog_ecommerce")
dbutils.widgets.text("schema","bronze")
```

#Constantes

```python
storageName = dbutils.widgets.get("storageName")
container = dbutils.widgets.get("container")
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
```

#Estructura

```python
product_schema = StructType(fields=[
    StructField("product_id", StringType(), False),
    StructField("product_category_name", StringType(), False),
    StructField("product_name_lenght", IntegerType(), False),
    StructField("product_description_lenght", IntegerType(), False),
    StructField("product_photos_qty", IntegerType(), False),
    StructField("product_weight_g", IntegerType(), False),
    StructField("product_length_cm", IntegerType(), False),
    StructField("product_height_cm", IntegerType(), False),
    StructField("product_width_cm", IntegerType(), False),
])
```

#Leer CSV

```python
df_productos = spark.read.format("csv").option('header', True).schema(product_schema).load(f"abfss://{container}@{storageName}.dfs.core.windows.net/productos")
```

#Guardar tabla

```python
df_productos.write.mode("overwrite").saveAsTable(f"{catalog}.{schema}.productos")
```
