#Import

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

#Widgets

```python
dbutils.widgets.removeAll()
dbutils.widgets.text("storageName", "adlsecommerceproject")
dbutils.widgets.text("container","raw")
dbutils.widgets.text("catalog","catalog_ecommerce")
dbutils.widgets.text("schema","bronze")
```

#Constantes

```python
storageName = dbutils.widgets.get("storageName")
container = dbutils.widgets.get("container")
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
```

#Estructura

```python
seller_schema = StructType(fields=[
    StructField("seller_id", StringType(), False),
    StructField("seller_zip_code_prefix", StringType(), False),
    StructField("seller_city", IntegerType(), False),
    StructField("seller_state", IntegerType(), False)
])
```

#Leer CSV

```python
df_seller = spark.read.format("csv").option('header', True).schema(seller_schema).load(f"abfss://{container}@{storageName}.dfs.core.windows.net/vendedores")
```

#Guardar en la Tabla

```python
df_seller.write.mode("overwrite").saveAsTable(f"{catalog}.{schema}.vendedores")
```
