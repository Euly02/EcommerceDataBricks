#Imports

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

#widgets

```python
dbutils.widgets.removeAll()
dbutils.widgets.text("storageName", "adlsecommerceproject")
dbutils.widgets.text("container","raw")
dbutils.widgets.text("catalog","catalog_ecommerce")
dbutils.widgets.text("schema","bronze")
```

#Constantes

```python
storageName = dbutils.widgets.get("storageName")
container = dbutils.widgets.get("container")
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
```

#Estructura

```python
items_schema = StructType(fields=[
    StructField("order_id", StringType(), False),
    StructField("payment_sequential", StringType(), False),
    StructField("payment_type", StringType(), False),
    StructField("payment_installments", StringType(), False),
    StructField("payment_value", StringType(), False),
])
```

#Leer csv

```python
df_payments = spark.read.format("csv").option('header', True).schema(items_schema).load(f"abfss://{container}@{storageName}.dfs.core.windows.net/payments")
```

#Guardar tabla

```python
df_payments.write.mode("overwrite").saveAsTable(f"{catalog}.{schema}.payments")
```
