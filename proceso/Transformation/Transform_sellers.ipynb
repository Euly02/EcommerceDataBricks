#Imports

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

#Widgets

```python
dbutils.widgets.removeAll()
dbutils.widgets.text("catalog","catalog_ecommerce")
dbutils.widgets.text("schema_source", "bronze")
dbutils.widgets.text("schema_sink", "silver")
```

#Constantes

```python
catalog         =  dbutils.widgets.get("catalog")
schema_source   =  dbutils.widgets.get("schema_source")
schema_sink     = dbutils.widgets.get("schema_sink")
```

#Preparacion

```python
df_sellers_bronze = spark.table(f"{catalog}.{schema_source}.vendedores")
```

#Conversion

```python
df_sellers_silver = df_sellers_bronze \
    .filter(col("seller_id").isNotNull()) \
    .withColumn("ingestion_date", current_timestamp()) \
    .dropDuplicates(["seller_id"])
```

#Guardar tabla

```python
df_sellers_silver.write.mode("overwrite").saveAsTable(f"{catalog}.{schema_sink}.vendedores")
```
