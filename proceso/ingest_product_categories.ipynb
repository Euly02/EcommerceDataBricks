#Imports

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

#Widgets

```python
dbutils.widgets.removeAll()
dbutils.widgets.text("storageName", "adlsecommerceproject")
dbutils.widgets.text("container","raw")
dbutils.widgets.text("catalog","catalog_ecommerce")
dbutils.widgets.text("schema","bronze")
```

#Constantes

```python
storageName = dbutils.widgets.get("storageName")
container = dbutils.widgets.get("container")
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
```

#Estructura

```python
categories_schema = StructType(fields=[
    StructField("product_category_name", StringType(), False),
    StructField("product_category_name_english", StringType(), False),
])
```

#Leer CSV

```python
df_categories = spark.read.format("csv").option('header', True).schema(categories_schema).load(f"abfss://{container}@{storageName}.dfs.core.windows.net/categories")
```

#Guardar tabla

```python
df_categories.write.mode("overwrite").saveAsTable(f"{catalog}.{schema}.categories")
```
