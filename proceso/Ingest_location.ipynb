#Imports

```python
from pyspark.sql.functions import *
from pyspark.sql.types import * 
```

#Widget


```python
dbutils.widgets.removeAll()
```

```python
dbutils.widgets.text("storageName", "adlsecommerceproject")
dbutils.widgets.text("container","raw")
dbutils.widgets.text("catalog","catalog_ecommerce")
dbutils.widgets.text("schema","bronze")
```

#Constantes

```python
storageName = dbutils.widgets.get("storageName")
container = dbutils.widgets.get("container")
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")
```

#Estructura

```python
location_schema = StructType(fields=[
    StructField("geolocation_zip_code_prefix", StringType(), False),
    StructField("geolocation_city", StringType(),False),
    StructField("geolocation_state", StringType(),False),      
])
```

#Leer CSV

```python
df_location = spark.read.option('header', True).schema(location_schema).csv(f"abfss://{container}@{storageName}.dfs.core.windows.net/location")
```

#Guardar Tabla

```python
df_location.write.mode("overwrite").saveAsTable(f"{catalog}.{schema}.location")
```
