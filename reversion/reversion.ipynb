###Revocar permisos a data analysts

```sql
USE CATALOG catalog_ecommerce;
REVOKE USAGE ON CATALOG catalog_ecommerce FROM `Data_analyst`;
REVOKE USAGE ON SCHEMA gold FROM `Data_analyst`;
REVOKE SELECT ON TABLE gold.delivery_by_states FROM `Data_analyst`;
REVOKE SELECT ON TABLE gold.sales_by_category FROM `Data_analyst`;
REVOKE SELECT ON TABLE gold.ventas_mensuales FROM `Data_analyst`;
```

###Revocar permisos a data scientific

```sql
REVOKE USAGE ON CATALOG catalog_ecommerce FROM `Data_scientifics`;
REVOKE USAGE ON SCHEMA bronze FROM `Data_scientifics`;
REVOKE SELECT, MODIFY ON SCHEMA bronze FROM `Data_scientifics`;
REVOKE USAGE ON SCHEMA silver FROM `Data_scientifics`;
REVOKE SELECT, MODIFY ON SCHEMA silver FROM `Data_scientifics`;
REVOKE USAGE ON SCHEMA gold FROM `Data_scientifics`;
REVOKE SELECT, MODIFY ON SCHEMA gold FROM `Data_scientifics`;
```
