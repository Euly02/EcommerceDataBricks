###Permisos a grupo de Data Analysts

```sql
USE CATALOG catalog_ecommerce;
GRANT USAGE ON CATALOG catalog_ecommerce TO `Data_analyst`;
GRANT USAGE ON SCHEMA gold TO `Data_analyst`;
GRANT SELECT ON TABLE gold.delivery_by_states TO `Data_analyst`;
GRANT SELECT ON TABLE gold.sales_by_category TO `Data_analyst`;
GRANT SELECT ON TABLE gold.ventas_mensuales TO `Data_analyst`;
```

###Permisos a grupo de Data scientifics

```sql
GRANT USAGE ON CATALOG catalog_ecommerce TO `Data_scientifics`;
GRANT USAGE ON SCHEMA bronze TO `Data_scientifics`;
GRANT SELECT, MODIFY ON SCHEMA bronze TO `Data_scientifics`;
GRANT USAGE ON SCHEMA silver TO `Data_scientifics`;
GRANT SELECT, MODIFY ON SCHEMA silver TO `Data_scientifics`;
GRANT USAGE ON SCHEMA gold TO `Data_scientifics`;
GRANT SELECT, MODIFY ON SCHEMA gold TO `Data_scientifics`;
```

```sql
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-bronze` to `Data_scientifics`;
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-silver` to `Data_scientifics`;
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-gold` to `Data_scientifics`;
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `exlt-raw` to `Data_scientifics`;
```
